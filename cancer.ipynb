# Load Library

```python
import numpy as np
import pandas as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer 
```

# Load Dataset

```python
bc = load_breast_cancer()
```

# Dataset Info

```python
bc.target.shape
bc.data.shape
```

# PreProccessing

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(bc.data , bc.target , test_size=0.2)
```

```python
print(f"feature size: {x_train.shape} and target size : {x_test.shape}")
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_train[0])
```

# Classification

## 1.Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train , y_train)
```

# Prediction

```python
y_pred_train_Bayes = gnb.predict(x_train)
y_pred_test_Bayes = gnb.predict(x_test)
```

# Score

```python
from sklearn.metrics import accuracy_score , recall_score ,precision_score
acc_train_bayes = accuracy_score(y_train , y_pred_train_Bayes)
print(f"Accuracy Score y_train is :{acc_train_bayes}")
acc_test_bayes = accuracy_score(y_test , y_pred_test_Bayes)
print(f"Accuracy Score y_test is :{acc_test_bayes}")
rcl_train_bayes = recall_score(y_train , y_pred_train_Bayes)
print(f"Recall Score y_train is :{rcl_train_bayes}")
rcl_test_bayes = recall_score(y_test , y_pred_test_Bayes)
print(f"Recall Score y_test is :{rcl_test_bayes}")
prc_train_bayes = precision_score(y_train , y_pred_train_Bayes)
print(f"Percision Score y_train is :{prc_train_bayes}")
prc_test_bayes = precision_score(y_test , y_pred_test_Bayes)
print(f"Percision Score y_train is :{prc_test_bayes}")
```

## 2.KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train , y_train)
```

# Prediction
